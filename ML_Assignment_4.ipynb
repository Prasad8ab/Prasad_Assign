1. What are the key tasks involved in getting ready to work with machine learning modeling?

The key tasks involved in getting ready to work with machine learning modeling are:

Data collection. The first step is to collect the data that you will be using to train your model. This data can come from a variety of sources, such as surveys, customer transactions, or sensor data.
Data preparation. Once you have collected your data, you need to prepare it for training your model. This may involve cleaning the data, removing outliers, and transforming the data into a format that your model can understand.
Model selection. There are many different machine learning models available, so you need to select the one that is right for your data and your problem.
Model training. Once you have selected a model, you need to train it on your data. This process involves feeding the data to the model and allowing it to learn the relationships between the features and the target variable.
Model evaluation. Once your model has been trained, you need to evaluate its performance on a held-out dataset. This will help you to determine how well the model will generalize to new data.
Model deployment. Once you are satisfied with the performance of your model, you can deploy it to production. This may involve making the model available as a web service or integrating it into your existing applications.

7. How do we deal with data outliers?
Outliers are data points that are significantly different from the rest of the data. They can be caused by a variety of factors, such as data entry errors, measurement errors, or the presence of unusual data points. Outliers can sometimes be removed from the data set, but this should only be done if they are clearly erroneous. In other cases, outliers can be treated as separate data points and analyzed separately.

There are a number of ways to deal with outliers. One common approach is to winsorize the data. This involves replacing the outliers with values that are within a certain distance of the nearest non-outlier data points. Another approach is to use robust statistical methods that are less sensitive to outliers.

8. What are the various central inclination measures? Why does mean vary too much from median in certain data sets?
The central inclination measures are statistical measures that are used to describe the central tendency of a data set. The most common central inclination measures are the mean, median, and mode.

The mean is the average of all the data points in a data set. The median is the middle value in a data set when the data points are arranged in ascending or descending order. The mode is the value that occurs most frequently in a data set.

The mean is sensitive to outliers, while the median is not. This is because the mean is calculated by adding up all the data points and dividing by the number of data points, while the median is calculated by finding the middle value in the data set. If there is an outlier in the data set, the mean will be pulled towards the outlier, while the median will not be affected.
