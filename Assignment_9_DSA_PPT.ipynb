```python
#Q1
def isPowerOfTwo(n):
    if n <= 0:
        return False
    while n > 1:
        if n % 2 != 0:
            return False
        n /= 2
    return True
```

```python
#Q2
def sumOfNaturalNumbers(n):
    return n * (n + 1) // 2
```

```python
#Q3
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
#Q4
def exponentiation(N, P):
    return N ** P
```

```python
#Q5
def findMax(arr):
    if len(arr) == 1:
        return arr[0]
    else:
        return max(arr[0], findMax(arr[1:]))
```

```python
#Q6
def nthTermOfAP(a, d, N):
    return a + (N - 1) * d
```

```python
#Q7
def permute(s):
    if len(s) == 1:
        return [s]
    else:
        permutations = []
        for i in range(len(s)):
            first_char = s[i]
            remaining_chars = s[:i] + s[i+1:]
            for perm in permute(remaining_chars):
                permutations.append([first_char] + perm)
        return permutations
```

```python
#Q8
def productOfArray(arr):
    product = 1
    for num in arr:
        product *= num
    return product
```
