1. What does one mean by the term "machine learning"?

Machine learning is a field of computer science that gives computers the ability to learn without being explicitly programmed. Machine learning algorithms use historical data as input to predict new output values.

2. Can you think of 4 distinct types of issues where it shines?

Machine learning shines in four distinct types of issues:

Classification: This is the task of assigning labels to data points. For example, a machine learning algorithm could be used to classify images as cats or dogs.
Regression: This is the task of predicting a continuous value. For example, a machine learning algorithm could be used to predict the price of a house based on its features.
Clustering: This is the task of grouping data points together based on their similarity. For example, a machine learning algorithm could be used to group customers together based on their purchasing habits.
Recommendation systems: This is the task of recommending products or services to users based on their past behavior. For example, a machine learning algorithm could be used to recommend movies to users based on the movies they have already watched.
3. What is a labeled training set, and how does it work?

A labeled training set is a set of data points that have been labeled with their correct output. For example, a labeled training set for a machine learning algorithm that classifies images as cats or dogs would consist of images of cats and dogs that have been labeled as such.

Machine learning algorithms learn by finding patterns in the labeled training set. Once the algorithm has learned these patterns, it can use them to predict the output for new data points that have not been labeled.

6. State the machine learning model that would be best to make a robot walk through various unfamiliar terrains?

The machine learning model that would be best to make a robot walk through various unfamiliar terrains is a reinforcement learning algorithm. Reinforcement learning algorithms learn by trial and error. The robot would be given a reward for walking successfully and a penalty for falling down. Over time, the robot would learn how to walk through different terrains without falling down.

8. Will you consider the problem of spam detection to be a supervised or unsupervised learning problem?

The problem of spam detection is a supervised learning problem. In supervised learning, the algorithm is given a labeled training set. In the case of spam detection, the training set would consist of emails that have been labeled as spam or not spam. The algorithm would then learn to identify spam emails based on the patterns in the labeled training set.

9. What is the concept of an online learning system?

An online learning system is a machine learning system that learns from data as it becomes available. In contrast, a traditional machine learning system learns from a static training set. Online learning systems are often used in situations where the data is constantly changing, such as in the case of spam detection.

10. What is out-of-core learning, and how does it differ from core learning?
Out-of-core learning is a type of online learning in which the algorithm is not allowed to access the entire training set at once. Instead, the algorithm is given a small subset of the training set at a time. This is useful in situations where the training set is too large to fit into memory.

12. What's the difference between a model parameter and a hyperparameter in a learning algorithm?

14. Can you name four of the most important Machine Learning challenges?
The four most important machine learning challenges are:
Data scarcity: There is often not enough data available to train a machine learning model.
Data bias: The data that is available may be biased, which can lead to the machine learning model making biased predictions.
Overfitting: The machine learning model may learn the training data too well, which can lead to the model making poor predictions on new data.
Underfitting: The machine learning model may not learn the training data well enough, which can also lead to the model making poor predictions on new data.

15. What happens if the model performs well on the training data but fails to generalize the results to new situations? Can you think of three different options?
If the model performs well on the training data but fails to generalize the results to new situations, then the model is overfitting the training data. There are three different options to address this problem:
Get more data.
Use a regularization technique.
Use a different machine learning algorithm.

16. What exactly is a test set, and why would you need one?
A test set is a set of data that is held out from the training data and is used to evaluate the performance of the machine learning model.

17. What is a validation set's purpose?
The purpose of a validation set is to evaluate the performance of the machine learning model during training.

18. What precisely is the train-dev kit, when will you need it, how do you put it to use?
The train-dev kit is a set of data that is used to train the machine learning model and to evaluate the performance of the model during training.
