Question 1: What is the concept of supervised learning? What is the significance of the name?

Answer: Supervised learning is a type of machine learning where the model is trained on a dataset of labeled data. This means that the data has been pre-classified, so the model knows what the correct output should be for each input. The name "supervised" comes from the fact that the model is being "supervised" by the labeled data.

Question 4: In supervised learning, what are classification and regression?

Question 6: Briefly describe the SVM model.

Answer: The SVM model is a supervised learning algorithm that can be used for both classification and regression tasks. It works by finding the hyperplane that best separates the data into two or more classes. The hyperplane is a line or a plane that has the maximum margin between the two classes. 

Question 7: In SVM, what is the cost of misclassification?

Answer: In SVM, the cost of misclassification is the penalty that is incurred when a data point is incorrectly classified. The cost of misclassification can be different for different classes, and it can be used to control the trade-off between accuracy and complexity of the model.

Question 8: In the SVM model, define Support Vectors.

Answer: In the SVM model, support vectors are the data points that lie on the hyperplane or closest to the hyperplane. These points are the most important for the model, and they are used to determine the position of the hyperplane.

Question 9: In the SVM model, define the kernel.

Answer: In the SVM model, the kernel is a function that is used to map the data into a higher dimensional space. This can be helpful in cases where the data is not linearly separable in the original space.

16. Explain the decision tree algorithm in a few words.

The decision tree algorithm is a supervised machine learning algorithm that can be used for both classification and regression tasks. It works by recursively partitioning the data into smaller and smaller subsets until each subset contains only data points of a single class. The decision tree is built by making a series of decisions, each of which is based on a single feature of the data.

17. What is the difference between a node and a leaf in a decision tree?

A node in a decision tree is a point where a decision is made. A leaf in a decision tree is a point where no further decisions can be made. The leaf nodes in a decision tree contain the class labels for the data points that fall into that leaf.

18. What is a decision tree's entropy?

The entropy of a decision tree is a measure of the impurity of the data at a node. A node with high entropy has data points that are evenly distributed across multiple classes. A node with low entropy has data points that are mostly concentrated in a single class.

19. In a decision tree, define knowledge gain.

Knowledge gain is a measure of how much information a decision split provides about the class labels of the data. A decision split with high knowledge gain indicates that the split is effective at separating the data points into different classes.

22. Briefly describe the random forest model.

A random forest is an ensemble learning algorithm that combines multiple decision trees to make predictions. Each decision tree in a random forest is trained on a random subset of the training data and a random subset of the features. This helps to reduce the variance of the individual decision trees and improve the overall performance of the random forest.
