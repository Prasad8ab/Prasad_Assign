```python
#Q1
def max_sum_subtree(root):
  """
  Finds the subtree with the maximum sum in the given binary tree.

  Args:
    root: The root node of the binary tree.

  Returns:
    The subtree with the maximum sum.
  """

  if not root:
    return 0

  max_sum = 0
  for child in (root.left, root.right):
    subtree_sum = max_sum_subtree(child)
    if subtree_sum > max_sum:
      max_sum = subtree_sum

  return max_sum + root.data
```

```python
#Q2
def construct_bst(arr):
  """
  Constructs a BST from the given level order traversal.

  Args:
    arr: The level order traversal of the BST.

  Returns:
    The root node of the BST.
  """

  queue = []
  for val in arr:
    node = TreeNode(val)
    queue.append(node)

    if len(queue) > 1:
      left_child = queue.pop(0)
      right_child = queue.pop(0)

      node.left = left_child
      node.right = right_child

  return queue[0]
```

```python
#Q3
def is_bst(arr):
  """
  Checks whether the given array can represent the level order traversal of a Binary Search Tree.

  Args:
    arr: The array to be checked.

  Returns:
    True if the array can represent the level order traversal of a Binary Search Tree, False otherwise.
  """

  is_bst = True
  for i in range(1, len(arr)):
    if arr[i] <= arr[i - 1]:
      is_bst = False
      break

    for j in range(i - 1, -1, -1):
      if arr[i] < arr[j]:
        is_bst = False
        break

  return is_bst
```
