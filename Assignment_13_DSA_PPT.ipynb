````python
#Q1
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def merge_lists(list1, list2):
    result = None
    current = None

    while list1 is not None and list2 is not None:
        if list1.value > list2.value:
            new_node = Node(list1.value)
            list1 = list1.next
        else:
            new_node = Node(list2.value)
            list2 = list2.next

        if result is None:
            result = new_node
            current = new_node
        else:
            current.next = new_node
            current = current.next

    while list1 is not None:
        new_node = Node(list1.value)
        current.next = new_node
        current = current.next
        list1 = list1.next

    while list2 is not None:
        new_node = Node(list2.value)
        current.next = new_node
        current = current.next
        list2 = list2.next

    return result
````

````python
#Q2
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def remove_duplicates(head):
    current = head

    while current is not None and current.next is not None:
        if current.value == current.next.value:
            current.next = current.next.next
        else:
            current = current.next

    return head
````

````python
#Q3
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def reverse_k_nodes(head, k):
    current = head
    prev = None
    next_node = None
    count = 0

    # Reverse the first k nodes
    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Connect the reversed part with the remaining list
    if next_node is not None:
        head.next = reverse_k_nodes(next_node, k)

    return prev
````

````python
#Q4
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def reverse_alternate_k_nodes(head, k):
    current = head
    prev = None
    next_node = None
    count = 0

    # Reverse the first k nodes
    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Connect the reversed part with the remaining list
    if next_node is not None:
        head.next = next_node

        # Skip the next k nodes
        for _ in range(k - 1):
            if current is None:
                break
            current = current.next

        if current is not None:
            next_head = reverse_alternate_k_nodes(current, k)
            head.next.next = next_head

    return prev
````

````python
#Q5
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def delete_last_occurrence(head, key):
    last = None
    to_delete = None
    current = head

    # Traverse the linked list and find the last occurrence of the key
    while current is not None:
        if current.value == key:
            last = to_delete
            to_delete = current
        current = current.next

    # If key is not found, return the original linked list
    if to_delete is None:
        return head

    # If the to_delete node is the head, update the head
    if last is None:
        head = head.next
    else:
        last.next = to_delete.next

    return head

````

````python
#Q6
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def merge_sorted_lists(a, b):
    dummy = Node()
    tail = dummy

    while a is not None and b is not None:
        if a.value <= b.value:
            tail.next = a
            a = a.next
        else:
            tail.next = b
            b = b.next
        tail = tail.next

    if a is not None:
        tail.next = a
    else:
        tail.next = b

    return dummy.next
````

````python
#Q7
class Node:
    def __init__(self, value=None):
        self.value = value
        self.prev = None
        self.next = None


def reverse_doubly_linked_list(head):
    current = head
    prev = None

    while current is not None:
        prev = current.prev
        current.prev = current.next
        current.next = prev
        current = current.prev

    if prev is not None:
        head = prev.prev

    return head
````
