Q1. Define the relationship between a class and its instances. Is it a one-to-one or a one-to-many partnership, for example?

A class and its instances have a one-to-many relationship. A class is a blueprint for creating objects, and each object is an instance of its class. For example, the class Dog can be used to create many different instances of dogs, each with its own unique attributes and behavior.

Q2. What kind of data is held only in an instance?

Instance data is data that is specific to a particular instance of a class. For example, the name of a dog is an instance data because it is unique to each individual dog. Class data, on the other hand, is data that is shared by all instances of a class. For example, the breed of dog is a class data because it is the same for all dogs of that breed.

Q3. What kind of knowledge is stored in a class?

A class stores knowledge about the attributes and behavior of its instances. Attributes are the data that an instance stores, such as its name, age, and breed. Behavior is the way that an instance interacts with the world, such as how it walks, barks, and eats.

Q4. What exactly is a method, and how is it different from a regular function?

A method is a function that is defined inside a class. Methods are used to perform actions on instances of the class. For example, the bark() method is a method that is defined in the Dog class. This method can be used to make any dog instance bark.

The main difference between a method and a regular function is that a method can access the instance data of the class that it is defined in. For example, the bark() method can access the name attribute of the dog instance that it is called on.

Q7. How do you distinguish between a class variable and an instance variable?

Class variables are defined outside of any method, and they are shared by all instances of the class. Instance variables are defined inside of a method, and they are unique to each instance of the class.

For example:
class Dog:
    species = "Canis lupus familiaris"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(self.name + " is barking!")

n this example, the species variable is a class variable. This means that it is shared by all instances of the Dog class. The name and age variables are instance variables. This means that they are unique to each instance of the Dog class.
