1. What is the relationship between def statements and lambda expressions ?
Def statements and lambda expressions are both ways to define functions in Python. However, there are some key differences between the two.

Def statements are more versatile than lambda expressions. They can take any number of arguments and can return any type of value. Lambda expressions, on the other hand, can only take one argument and can only return a single value.
Def statements are more explicit than lambda expressions. They require the programmer to specify the function name, the argument list, and the return type. Lambda expressions, on the other hand, can be defined inline, without any of these details.
In general, def statements should be used when you need to define a complex function that will be used multiple times in your program. Lambda expressions should be used when you need to define a simple function that will only be used once.

2. What is the benefit of lambda?

The main benefit of lambda expressions is that they are concise and easy to read. They can be used to define small, one-off functions that would otherwise be difficult to define using def statements.

For example, the following lambda expression defines a function that takes an integer and returns its square:
square = lambda x: x * x

This function can be used just like any other function:
print(square(4))
The output of this code is 16.

3. Compare and contrast map, filter, and reduce.

Map, filter, and reduce are all built-in functions in Python that can be used to iterate over a sequence of elements and perform a specific operation on each element.

Map takes a function and a sequence of elements and applies the function to each element in the sequence. The result is a new sequence of elements, each of which is the result of applying the function to the corresponding element in the original sequence.
Filter takes a function and a sequence of elements and returns a new sequence of elements that only contains the elements for which the function returns True.
Reduce takes a function and a sequence of elements and repeatedly applies the function to the elements in the sequence, starting with the first two elements. The result of the reduce function is a single value.
Here is an example of how to use each of these functions:


```python
numbers = [1, 2, 3, 4, 5]

# Map
doubled_numbers = map(lambda x: x * 2, numbers)
print(doubled_numbers)
# [2, 4, 6, 8, 10]

# Filter
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(even_numbers)
# [2, 4, 6]

# Reduce
product = reduce(lambda x, y: x * y, numbers)
print(product)
# 120
```

5. What are recursive functions, and how are they used?

Recursive functions are functions that call themselves. They can be used to solve problems that involve self-similar structures, such as trees and fractals.

For example, the following recursive function calculates the factorial of a number:

def factorial(n: int) -> int:
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)

This function works by first checking if the number n is equal to 0. If it is, the function returns 1. Otherwise, the function returns the value of n multiplied by the factorial of n - 1.

Recursive functions can be powerful tools, but they can also be difficult to understand and debug. It is important to use them carefully and only when they are necessary.

6. What are some general design guidelines for coding functions?

here are some general design guidelines for coding functions:

Give your functions clear and descriptive names. The name of a function should give a good indication of what the function does.
Use meaningful arguments. The arguments to a function should be named in a way that makes it clear what they represent.
