In the second year, you will earn interest on your original investment of $100, as well as on the interest you earned in the first year, which is $0.50. So, you will earn $5.25 in interest in the second year. This process continues over time, and the amount of interest you earn grows exponentially.

A higher rate of interest that does not compound will still earn you more interest than a lower rate of interest, but the difference will not be as great as with compound interest. Q3. What is a histogram, exactly? Name a numpy method for creating such a graph.

A histogram is a graphical representation of the distribution of data. It shows the number of data points that fall within each range of values. For example, a histogram of the heights of a group of people might show that there are more people with heights between 5 feet and 5 feet 6 inches than there are people with heights between 5 feet 6 inches and 6 feet.

The numpy.histogram() function can be used to create a histogram. This function takes two arguments: the data and the number of bins. The data is the array of data points that you want to plot. The number of bins is the number of ranges that you want to divide the data into.

Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot product, outer product, and regular multiplication of two numpy arrays.

The three types of array multiplication between two numpy arrays are:

Dot product: The dot product is the sum of the products of the corresponding elements of two arrays. For example, if you have two arrays, a and b, with the following dimensions:

a = [1, 2, 3]
b = [4, 5, 6]

The dot product of a and b is:

a @ b = 30

Outer product: The outer product is the product of two arrays, where each element of the resulting array is the product of the corresponding elements of the two input arrays. For example, if you have two arrays, a and b, with the following dimensions:

a = [1, 2, 3]
b = [4, 5, 6]

The outer product of a and b is:

a[:, None] * b[None, :] = [[1 * 4, 1 * 5, 1 * 6], [2 * 4, 2 * 5, 2 * 6], [3 * 4, 3 * 5, 3 * 6]]

Regular multiplication: The regular multiplication of two arrays is the element-wise multiplication of the two arrays.
