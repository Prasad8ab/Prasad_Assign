Q10. What is the best way to use a Boolean array to mask another array?

The best way to use a Boolean array to mask another array is to use the where function. The where function takes two arguments: a Boolean array and an array. The where function will return a new array that contains the elements of the second array where the corresponding elements of the first array are True.

For example, the following code will create a new array that contains only the even numbers from the array A:

A = np.arange(10)
mask = A % 2 == 0
new_array = np.where(mask, A, -1)

This code will first create a Boolean array mask that contains True for even numbers and False for odd numbers. Then, it will create a new array new_array that contains the elements of the array A where the corresponding elements of the array mask are True. In this case, the new array will contain only the even numbers from the array A.
