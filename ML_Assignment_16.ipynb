Question 1: In a linear equation, what is the difference between a dependent variable and an independent variable?

Answer:

Dependent variable: The variable that is being predicted.
Independent variable: The variable that is used to do the predicting.
Question 2: What is the concept of simple linear regression? Give a specific example.

Answer:

Simple linear regression: A statistical method that uses a straight line to model the relationship between two variables.
Example: The relationship between the height of a child and their age. As a child grows older, their height will increase. This relationship can be modeled by a straight line, with the independent variable being the child's age and the dependent variable being their height.
Question 3: In a linear regression, define the slope.

Answer:

Slope: The amount that the dependent variable changes for every one unit change in the independent variable.

Slope: 0.5
Question 5: In linear regression, what are the conditions for a positive slope?

Answer:

The independent variable and the dependent variable must be positively correlated. This means that as the independent variable increases, the dependent variable must also increase.
The independent variable must not be perfectly correlated with the dependent variable. This means that there must be some variation in the dependent variable that cannot be explained by the independent variable.
Question 6: In linear regression, what are the conditions for a negative slope?

Answer:

The independent variable and the dependent variable must be negatively correlated. This means that as the independent variable increases, the dependent variable must decrease.
The independent variable must not be perfectly correlated with the dependent variable. This means that there must be some variation in the dependent variable that cannot be explained by the independent variable.
Question 7: What is multiple linear regression and how does it work?

Answer:

Multiple linear regression: A statistical method that uses multiple independent variables to predict a dependent variable.
How it works: Finds the values of the coefficients that minimize the sum of the squared residuals. The residuals are the differences between the actual values of the dependent variable and the predicted values of the dependent variable. The coefficients are found using a process called ordinary least squares.
Question 8: In multiple linear regression, define the number of squares due to error.

Answer:

Number of squares due to error: The sum of the squared differences between the actual values of the dependent variable and the predicted values of the dependent variable.
Measure of how well the model fits the data. A lower number of squares due to error indicates a better fit.

Question 9: In multiple linear regression, define the number of squares due to regression.

Answer:

Number of squares due to regression: The sum of the squared differences between the mean of the dependent variable and the predicted values of the dependent variable.
Measure of how well the model explains the variation in the dependent variable. A higher number of squares due to regression indicates a better fit.

Question 10: What is multicollinearity?

Answer:

Multicollinearity: A condition in which two or more independent variables are highly correlated with each other.
Can cause problems with multiple linear regression, because it can make it difficult to determine which independent variable is actually affecting the dependent variable.

Question 11: What is heteroskedasticity?

Answer:

Heteroskedasticity: A condition in which the variance of the residuals is not constant across all values of the independent variable.
Can cause problems with multiple linear regression, because it can make it difficult to estimate the standard errors of the coefficients.

Question 12: Describe the concept of ridge regression.

Answer:

Ridge regression: A statistical method that is used to reduce the effects of multicollinearity in multiple linear regression.
Does this by adding a penalty to the sum of the squared coefficients. This penalty makes the coefficients smaller, which reduces the correlation between them.

Question 13: Describe the concept of lasso regression.

Answer:

Lasso regression: A statistical method that is also used to reduce the effects of multicollinearity in multiple linear regression.
Does this by adding a penalty to the sum of the absolute values of the coefficients. This penalty makes some of the coefficients zero, which eliminates the independent variables that are not important for predicting the dependent variable.

Question 14: What is polynomial regression?

Answer:

Polynomial regression: A statistical method that uses a polynomial function to model the relationship between two variables.
The degree of the polynomial function determines how many independent variables are used to model the relationship. For example, a quadratic polynomial function would use two independent variables to model the relationship.

Question 15: Describe the basis function.

Answer:

Basis function: A function that is used to transform the independent variable into a new space where the relationship between the independent variable and the dependent variable is easier to model.
For example, a common basis function is the polynomial function.

Question 16: Describe how logistic regression works.

Answer:

Logistic regression: A statistical method that is used to predict the probability of a binary outcome.
The independent variable is plotted on the x-axis and the probability of the outcome is plotted on the y-axis.
The model is trained on a dataset of known outcomes and then used to predict the probability of the outcome for new data.
